### Data Cleaning
##### Not all of the columns are needed for analysis, so another dataset is created, a subset of the original data frame. This is labeled breastcancer, and the columns are renamed. This allows for greater readability and easy recognition during analysis.


```python
breastcancer = df[['YearStart', 'YearEnd', 'LocationAbbr', 'LocationDesc', 'Question', 'DataValueUnit', 'DataValueType', 'DataValue', 'StratificationCategory1', 'Stratification1']]
breastcancer
```

#### Not all of the responses within the selected columns are necessary. 
#### Only the values for the data ending in 2020 will be used (as that is the year that data for breast cancer mortality rates and mammography rates have in common).
#### For accurate comparison between states, crude rates, as opposed to number of cases reported will be used. To ensure this, the data is filtered, set to only include the rows that contain crude. 
#### Questions used will center around breast cancer and mammography, so those key words are filtered on as well.
#### The groups of interest are overall (total rate for each state), Black, White and Hispanic groups, so a filter is applied to exclude all entries that do not fall into those categories.
#### There are also some locations in the set are not states, such as Guam. (District of Columbia and an overall value for the US are also included, but these will be kept, as they may be useful during analysis.)
#### An new data set will be created to represent the updated dataset. It will be called breastcancer_filtered.

```python
breastcancer_filtered = breastcancer[
    (breastcancer["YearEnd"].astype(str).str.contains("2020", case=False, na=False)) &
    (breastcancer["DataValueType"].str.contains("Crude", case=False, na=False)) &
    (~breastcancer["Stratification1"].str.contains("American Indian|Alaska Native|Asian|Pacific Islander|Multiracial|Hawaiian|Male|Female", case=False, na=False)) &
    (~breastcancer["LocationAbbr"].str.contains("GU|VI|PR", case=False, na=False)) &
    ((breastcancer["Question"].str.contains("Breast Cancer", case=False, na=False)) |
     (breastcancer["Question"].str.contains("Mammography", case=False, na=False)))
]
breastcancer_filtered
```

```python
pd.options.mode.copy_on_write = True
```

#### For easier viewing, some columns will be renamed.

```python
breastcancer_filtered.rename(columns={'YearStart':'year_start'}, inplace=True)
breastcancer_filtered.rename(columns={'YearEnd': 'year_end' },inplace=True)
breastcancer_filtered.rename(columns={'LocationAbbr':'abbr'}, inplace=True)
breastcancer_filtered.rename(columns={'LocationDesc':'location'},inplace=True) 
breastcancer_filtered.rename(columns={'Question':'question'}, inplace=True)
breastcancer_filtered.rename(columns={'DataValueUnit':'data_unit' },inplace=True)
breastcancer_filtered.rename(columns={'DataValueType':'data_type'}, inplace=True)
breastcancer_filtered.rename(columns={'DataValue':'data_value'},inplace=True) 
breastcancer_filtered.rename(columns={'StratificationCategory1': 'category'}, inplace=True)
breastcancer_filtered.rename(columns={'Stratification1': 'classification'}, inplace=True)
```

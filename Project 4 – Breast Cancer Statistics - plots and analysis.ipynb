# Research Question 1 
## Do some states have a higher mammography rate than others?

#### The first research question requires only the use of the question "Mammography use among women aged 50-74 years" for each state. As this data is being viewed at a state level, the value for the category overall will be used. 
#### To facilitate this, a subset containing only the necessary columns will be created. 

```python
dataset1 = breast_cancer_data[
        ((breast_cancer_data["classification"].str.contains("Overall", case=False, na=False)) &
        (breast_cancer_data["question"].str.contains("Mammography", case=False, na=False)))
]
dataset1
```

#### To ensure that only the fifty states needed are in the data set, the shape function is used.
#### Once this is verified, the data can be analyzed by looking at the mammography rates across the four regions.

```python
dataset1.shape
```

```python
plt.figure(figsize=(16, 8))
ax= sns.barplot(
    data=dataset1,
    x='region',
    y='data_value',
    hue='region',
    order=["Northeast", "Midwest", "South", "West"], 
    legend=False,
    palette='colorblind',
    errorbar=None)
plt.title('Mammography Use by Region')
plt.xlabel('Region'), 
plt.ylabel('Mammography Use')

for bar in ax.patches:
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # this positions the label at the center of each bar
        bar.get_height(),  # positions the label at the top of each bar
        f"{bar.get_height():.1f}",  # limits the value to only one decimal place
        ha='center', va='bottom', fontsize=10, color="black"
    )

plt.savefig("mammography_use_by_region.svg", format="svg", dpi=300, bbox_inches="tight")

plt.show()
```

#### Looking at the mammography use by state, rates are reported to be very similar across each region with the Northeast taking a slight lead while the West dips a little behind.
#### To fully answer the research question however, a deeper look is needed to see if this consistency continues at state level.
#### The easiest way to visualize this is by looking at the states within each region to see if there are significant differences within regions.

Usage is generally consistent between the states in each region, with the most variation occurring in the West which has the states with the four lowest use rates. #### From questions 1 and 2 we know that mammography use and mortality rates are close across the four regions. For easy comparison, these variables can be plotted side by side, by region.
#### To do this a column, 'type', will be added to each dataset, before being combined under the name dataset3.

```python
dataset1 = dataset1.copy()
dataset2 = dataset2.copy()

dataset1.loc[:, 'type'] = 'Mammography Use'
dataset2.loc[:, 'type'] = 'Mortality'

dataset3 = pd.concat([dataset1, dataset2])

dataset3
```

```python
plt.figure(figsize=(16, 8))
ax = sns.barplot(
    data=dataset3,
    x='region',
    y='data_value',
    hue='type',  # differentiates between mammography use and mortality
    order=["Northeast", "Midwest", "South", "West"],  # sets region order
    palette=['#0173B2', '#D55E00'],  # sets colors for chart
    errorbar=None
)

for bar in ax.patches:
    ax.text(
        bar.get_x() + bar.get_width() / 2,  
        bar.get_height(),  
        f"{bar.get_height():.1f}",  
        ha='center', va='bottom', fontsize=10, color="black"
    )

plt.title('Mammography Use and Breast Cancer Mortality by Region')
plt.xlabel('Region')
plt.ylabel('Rate')
plt.legend(title="Type")  # shows legend for type

plt.savefig("mammography_use_and_breast_cancer_mortality_by _region.svg", format="svg", dpi=300, bbox_inches="tight")

plt.show()
```
